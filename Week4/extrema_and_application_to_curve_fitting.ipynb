## Exercise 1

*Locate the local maximum, local minimum and saddle points of $f(x,y)=2(x^2-y^2)-x^4+y^4$.*

```python
x, y, z = var('x y z')

f(x,y) = 2*(x**2-y**2)-x**4+y**4

S = implicit_plot3d(z-f(x,y), (x,-2,2), (y,-2,2), (z,-1,1), color='red', opacity=0.8, plot_points=200)

#show(S)
```

```python
ct = contour_plot(f(x,y), (x,-2,2), (y,-2,2), contours=20, fill=False, figsize=6, cmap='winter')

show(ct)
```

```python
def classify_extreme(f,a,b):
    f11=diff(f,x,x)(a,b)
    f22=diff(f,y,y)(a,b)
    f12=diff(f,x,y)(a,b)
    D=f11*f22-f12^2
    if(D>0):
        if(f11>0):
            return "local minimum"
        else:
            if(f11<0):
                return "local maximum"
            else:
                return "inconclusive"

    else:
        if(D<0):
            return "saddle point"
        else:
            if(D==0):
                return "inconclusive"
```

```python
fx = f.diff(x)
fy=f.diff(y);
fxx=f.diff(x,x)
fyy=f.diff(y,y)
fxy=f.diff(x,y)
fyx=f.diff(y,x)
```

```python
data = [["Critical Points ", " Type", 'Values']]
s=solve([fx==0,fy==0],[x,y],solution_dict=True)
for sol in s:
    if ((sol[x] in RR) and (sol[y] in RR)):
        a=sol[x]
        b=sol[y]
        data.append([(sol[x],sol[y]), classify_extreme(f,a,b),f(x=a,y=b)])
table(data,frame=True,header_row=True)
```

## Exercise 6

*Fit a quadratic $y=ax**2+bx+c$ to a set of point :*

$(0, 5),(1, 7),(2, 3),(3, 5),(4, 14),(5, 31),(6, 50),(7, 75),(8, 100),(9, 125)$

```python
xx = range(10)
yy = [-5, -7, -2, 5, 14, 31, 50, 75, 100, 125]

data = [(xx[i], yy[i]) for i in range(10)]
```

```python
a, b, c, t = var('a b c t')

err = [yy[i]-(a*xx[i]**2+b*xx[i]+c) for i in range(10)]

f(a, b, c) = sum([err[i]**2 for i in range(10)])
```

```python
fa = f.diff(a)
fb = f.diff(b)
fc = f.diff(c)

eqns = [fa==0, fb==0, fc==0]

sols = solve(eqns, [a,b,c], solution_dict=True)

show(sols[0])
```

```python
p = plot(sols[0][a]*t**2+sols[0][b]*t+sols[0][c], (t, -1, 10))
pts = points(data,size=20,color='red',figsize=4)

show(p+pts)
```

## Exercise 7

*Fit the best fit line $y = ax + b$ to the set of points :*

$(1, 1),(2, 1),(3, 5),(4, 7),(5, 12),(6, 15),(7, 20),(8, 19),(9, 25),(10, 24)$

```python
xx = range(1,11)
yy = [-1, 1, -5, -7, -12, -15, -20, -19, -25, -24]

data = [(xx[i], yy[i]) for i in range(10)]
```

```python
a, b, t = var('a b t')

err = [yy[i]-(a*xx[i]+b) for i in range(10)]

f(a, b, c) = sum([err[i]**2 for i in range(10)])
```

```python
fa = f.diff(a)
fb = f.diff(b)


eqns = [fa==0, fb==0]

sols = solve(eqns, [a,b], solution_dict=True)

show(sols[0])
```

```python
p = plot(sols[0][a]*t+sols[0][b], (t, 0, 11))
pts = points(data,size=20,color='red',figsize=4)

show(p+pts)
```
