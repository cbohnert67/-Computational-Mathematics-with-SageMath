## Exercise 1

*Find the average value of $f(x) = \frac{x}{(1+x^2)^2}$ on $[0, 2]$. Hence verify the mean value theorem of
integral. Also plot the graph of f and the rectangle whose area is same as the area under thecurve.*

```python
x = var('x')

f(x) = x/(1+x**2)**2

a, b = 0, 2

Favg = 1/(b-a)*integral(f(x),x,a,b)

print(f'Favg = {Favg}')

c1 = find_root(f(x)==Favg,0,0.5); print(f'c1 = {c1}')
print(f'f(c1) = {f(c1)}')
c2 = find_root(f(x)==Favg,1,1.5); print(f'c2 = {c2}')
print(f'f(c2) = {f(c2)}')

f.plot(a,b,figsize=4)+plot(Favg,a,b,color='red',fill=True)
```

## Exercise 2

*The linear density in a rod $12$ m long is $\frac{18}{\sqrt{2x+1}} kg/m$ where $x$ is measured in meters from 
one end of the rod. Find the average density of the rod.*

```python
linear_density(x) = 18/sqrt(2*x+1)

L = 12

avg_density = 1/L*linear_density.integral(x, 0, L)

print(f'Average density of the rod : {avg_density}')
```

## Exercise 3

*Evaluate :*

(i) $\int_{-\infty}^{\infty} \frac{1}{1+x^{2}}, d x$

(ii) $\int_{0}^{\infty} e^{-x^{2}} x^{2 n-1} d x$

```python
# i 

f(x) = 1/(1+x**2)

I = f.integrate(x, -infinity, infinity)

print(f'I = {I}')
```

```python
# ii

n = var('n')

assume(2*n+1>0)

f(x) = exp(-x**2)*x**(2*n*1)

J = f.integrate(x, 0, infinity)

print(f'J = {J}')
```

*Consider the function $f(x)=\frac{1}{\sqrt{2 \pi} s} e^{-\frac{(x-m)^{2}}{2 s^{2}}}$, where $s > 0$ and $m$ is any real number. Plot the graph of $f$ for $m = 5$, $s = 2$. Show that $\int_{-\infty}^{\infty} f(x) d x=1$.*

```python
s, m = var(' s m')

f(x, m, s)=1/(sqrt(2*pi)*s)*exp(-(x-m)**2/(2*s**2))

p = plot(f(x, m=5, s=2), (x, -3,13), fill=True)

show(p)
```
