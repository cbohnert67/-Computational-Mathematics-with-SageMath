## Exercise 1

*Graph the curve $y = (1+x^2)^{3/2}$ for $0 \leq x \leq 4$ and hence find its arc length.*

```python
x = var('x')

f(x) = (1+x^2)^(3/2)

p = plot(f(x), (x,0,4), color='red', thickness=2, axes_labels=['$x$', '$f(x)$'])

show(p)
```

```python
dydx = diff(f(x), x)

l(x) = sqrt(1+dydx**2)

L = numerical_integral(l(x), 0, 4)

print(f'Arc length of f is {L[0]}')
```

## Exercise 2

*Find the arc length of the curve $y^3 = (1+2x)^2$ for $0 \leq x \leq 1$, $ y > 0$.*

```python
x, y = var('x y')

f(x,y) = y**3-(1+2*x)**2

l(x) = sqrt(1+f.implicit_derivative(y,x)**2)

l(x)=l.substitute(y=(1+2*x)**(2/3))

L = numerical_integral(l(x), 0, 1)

print(f'Arc length of f is {L[0]}')
```

## Exercise 4

*Find the volume of the solid of revolution of the curve $y = \cos(x)$, $0 \leq x \leq \pi$ revolve about the axis $y = 1$.*

```python
x = var('x')

f(x) = cos(x)-1

V = integral(pi*f(x)^2,x,0,pi)

print(f'Volume of the solid is {V.n()}')
```

## Exercise 5

*Let $R$ be the region in the first quadrant bounded by the curves $y = x ^2$ and $y = x âˆ’ 3x^3$.*

*Calculate (i) the area of $R$ (ii) the volume obtained by rotating $R$ about the x-axis. and (iii) the volume obtained by rotating $R$ about the y-axis.*

```python
f(x) = x**2
g(x) = x - 3*x**3

xmax = solve(f(x)-g(x), x)[1].rhs()

p = plot([f(x), g(x)], x, 0, xmax, color='red', thickness=2, axes_labels=['$x$', '$y$'], fill={0:[1]})

show(p)
```

```python
# i

A = integral(g(x)-f(x), x, 0, xmax)

print(f'Area between curve f and curve g is  {A.n()}')
```

```python
# ii

Vx = pi*integral(g(x)**2-f(x)**2, x, 0, xmax)

print(f'Volume obtained by rotating R about the x-axis is  {Vx.n()}')
```

Remark of iii.

Shape of region R is preversed whether we rotate it about x-axis or y-axis. Therefore, area of R is conserved by rotation, and hence the volume of the solid of revolution obtained by rotation about the y-axis is the same as $Vx$.
