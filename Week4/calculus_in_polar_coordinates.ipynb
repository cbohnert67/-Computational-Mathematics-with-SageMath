# Calculus in polar coordinates

## Exercise 1

*Find the area bounded by one loop of the curve $r=2 \cos(2 \theta)$, $0 \leq \theta \leq 2 \pi$.*

```python
theta = var('theta')

r(theta) = 2*cos(2*theta)

polar_plot(r(theta), 0, 2*pi,color='red', figsize=6, fill=True, thickness=3)
```

```python
A = integral(1/2*r(theta)^2,theta, 0, 2*pi)

print(f'Area bounded by one loop of the curve is {A.n()}')
```

## Exercise 2

*Find the arc length of the curve $r = 1+2 \cos (\theta)$, $0 \leq \theta \leq 2 \pi$.*

```python
theta = var('theta')

r(theta) = 1+2*cos(theta)

polar_plot(r(theta), 0, 2*pi,color='red', figsize=6,  thickness=3)
```

```python
dr = r.diff()

L = integral(sqrt(r^2+dr^2),theta,0,2*pi)

print(f'Length  of the curve is {L.n()}')
```

## Exercise 3

*Find the area that the curve $r=3*(1-cos(2 \theta))$, $0 \leq \theta \leq 2 \pi$ encloses.*

```python
theta = var('theta')

r(theta) = 3*(1-cos(2*theta))

polar_plot(r(theta), 0, 2*pi,color='red', figsize=6, fill=True, thickness=3)
```

```python
A = integral(1/2*r(theta)^2,theta, 0, 2*pi)

print(f'Area bounded by one loop of the curve is {A.n()}')
```

## Exercise 4

*The area of the surface generated by rotating the polar curve $r = f(\theta)$, $a \leq \theta \leq b$, with $0 \leq a < b \leq \pi$ about the x-axis is given by*

$$S=\int_{a}^{b} 2 \pi r \sin \theta \sqrt{r^{2}+\left(\frac{d r}{d \theta}\right)^{2}} d \theta$$

*Hence find the area of the surface generated by rotating the polar curve $r = \theta ^2$, $0 \leq \theta \leq \pi$ about the x-axis.*

```python
theta = var('theta')

r(theta) = theta**2

polar_plot(r(theta), 0, 2*pi,color='red', figsize=6, fill=True, thickness=3)
```

```python
dr = r.diff()

A = integral(2*pi*r(theta)*sin(theta)*sqrt(r(theta)**2+dr**2),theta,0,pi)

print(f'Area bounded by one loop of the curve is {A.n()}')
```
