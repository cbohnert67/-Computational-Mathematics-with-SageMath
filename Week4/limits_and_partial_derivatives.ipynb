## Exercise 1

*Show that $\lim _{(x, y) \rightarrow(0,0)} \frac{\sin \left(x^{3}+y^{3}\right)}{x^{2}+y^{2}}$ exists and hence find the same.*

```python
x, y, r, theta = var('x y r theta')

f(x,y) = sin(x**3+y**3)/(x**2+y**2)

g(r,theta) = f(x,y).substitute(x = r*cos(theta), y = r*sin(theta)).trig_simplify()

show(g)
```

```python
limit(g(r, theta), r=0)
```

## Exercise 5

*Show that the function $u=\frac{1}{\sqrt{x^{2}+y^{2}+z^{2}}}$ satisfies the Laplace equation $u_{x x}+u_{y y}+u_{z z}=0$.*

```python
x, y, z = var('x y z')

u(x,y,z) = 1/sqrt(x**2+y**2+z**2)

uxx = u.diff(x, 2)
uyy = u.diff(y, 2)
uzz = u.diff(z, 2)

factor(uxx+uyy+uzz)
```

## Exercise 6

*Plot the surface $z = x^2+xy+3y^2$ along with tangent plane at the point $(1, 1, 5)$.*

```python
x, y = var('x y')

f(x,y) = x**2 + x*y + 3*y**2

t(x,y) = f.diff(x)(1,1)*(x-1)+f.diff(y)(1,1)*(y-1)+f(1,1)

p1 = plot3d(f(x,y), (x,-2,2), (y, -2,2), color='red')
p2 = plot3d(t(x,y), (x,-2,2), (y, -2,2), color='blue')

show(p1+p2)
```

## Exercise 7

*If $z=x^2 y +3xy^4$, with $y=\sin(2t)$, $y=\cos(t)$. Find $\frac{dz}{dt}$ at $t=0$.*

```python
x, y, t = var('x y t')

f(x,y) = x**2*y +  3*x*y**4
g(t) = sin(2*t)
h(t) = cos(t)

dfdt = f.diff(x)*g.diff(t)+f.diff(y)*h.diff(t)
dfdt(t) = dfdt.substitute(x=sin(2*t), y=cos(t)).trig_simplify()


show(dfdt)
show(dfdt(0))
```

## Exercise 8

*Find the directional derivative of the function $f(x,y) = x^2y^3-4y$ at the point $(2, 1)$ in the direction of the vector $(2, 5)$.*

```python
x, y = var('x y')

f(x,y) = x**2*y**3 -4*y

v = f.gradient()(2,-1)
v.dot_product(vector([2,-5]))
```

## Exercise 9

*The Temperature at a point $(x, y, z)$ is given by $T(x, y, z)=150 e^{-x^{2}-2 y^{2}-5 z^{2}}$ where $T$ is measured in degree celsius, $x, y, $z are measured in meters.*

i. *Find the rate of change of $T$ at $P(2, 1, 2)$ in the direction of $(1, 1, 1)$.*

ii. *Find the direction $v$ in which the temperature increases with maximum speed at $P$.*

iii. *Find the maximum rate of increase in $T$ at $P$.*

```python
x, y, z = var('x y z')

T(x, y, z) = 150*exp(-x**2-2*y**2-5*z**2)
```

```python
# i

g = T.gradient()(2,-1, 2)

rate_of_change = g.dot_product(vector([1,-1, 1]))

rate_of_change.n()
```

```python
# ii

g.n()
```

```python
# iii

max_rate_of_change = g.dot_product(g)

max_rate_of_change.n()
```
