## Exercise 1

i. *Find the 1st four derivative of $f(t) = \ln(1 + t^2)$ and plot them along with the graph of $f$.*

ii. *What is the n-th deriavative of $x^x$ for various values of $n$?*

iii. *Find the first four derivatives of $f(x)=\sin(3x)+e^{-2x}+\ln(7x)$*

```python
# i

t = var('t')

f(t) = log(1+t**2)

df1(t) = diff(f(t),t)
df2(t) = diff(f(t), t, 2)
df3(t) = diff(f(t), t, 3)
df4(t) = diff(f(t), t, 4)

show(LatexExpr(r'f^{(1)}(t) = '),df1(t))
show(LatexExpr(r'f^{(2)}(t) = '),df2(t))
show(LatexExpr(r'f^{(3)}(t) = '),df3(t))
show(LatexExpr(r'f^{(4)}(t) = '),df4(t))

p = plot(f(t), (t, -2,2), color='black', legend_label=r'$f$')
p1 = plot(df1(t), (t, -2,2), color='red', legend_label=r'$f^{(1)}$')
p2 = plot(df2(t), (t, -2,2), color='green', legend_label=r'$f^{(2)}$')
p3 = plot(df3(t), (t, -2,2), color='blue', legend_label=r'$f^{(3)}$')
p4 = plot(df4(t), (t, -2,2), color='orange', legend_label=r'$f^{(4)}$')

show(p+p1+p2+p3+p4, legend_loc=(0.7,0.2))
```

```python
# ii

x = var('x')

f(x) = x**x

df = lambda x, n: diff(f(x), x, n)

show(LatexExpr(r'f^{(1)}(x) = '),df(x,1))
show(LatexExpr(r'f^{(3)}(x) = '),df(x, 3))
show(LatexExpr(r'f^{(5)}(x) = '),df(x, 5))
show(LatexExpr(r'f^{(10)}(x) = '),df(x, 10))
```

```python
# iii

x = var('x')

f(x) = sin(3*x)+exp(-2*x)+log(7*x)

df = lambda x, n: diff(f(x), x, n)

show(LatexExpr(r'f^{(1)}(x) = '),df(x,1))
show(LatexExpr(r'f^{(2)}(x) = '),df(x, 2))
show(LatexExpr(r'f^{(3)}(x) = '),df(x, 3))
show(LatexExpr(r'f^{(4)}(x) = '),df(x, 4))
```

## Exercise 3

*Consider the function $f(x)=e^{(x-1)^2}+\sin(\cos(5x))$. Verify the Lagrange mean value theorem for for $f$ in the interval $[1, 2]$. Plot the graph to explain the geometric meaning.*

```python
x = var('x')

f(x) = exp((x-1)**2)+sin(cos(5*x))
a,b = 1, 2

pf=plot(f,a,b)
cord=line([(a,f(a)),(b,f(b))],color='green',linestyle='--')
show(pf+cord,figsize=4)
```

```python
m=(f(b)-f(a))/(b-a) # slope of the chord
c1=find_root(f.diff(x)-m,1,1.3)
c2=find_root(f.diff(x)==m,1.6,2)
print(f'First value : {c1}')
print(f'Second value : {c2}')
```

```python
dfdx=f.diff()
l1(x)=f(c1)+dfdx(c1)*(x-c1)
l2(x)=f(c2)+dfdx(c2)*(x-c2)

pf=plot(f,a,b)
cord=line([(a,f(a)),(b,f(b))],color='green',linestyle='--')
t1=plot(l1(x),c1-0.2,c1+0.2,color='red')
t2=plot(l2(x),c2-0.2,c2+0.2,color='red')

p=pf+cord+t1+t2
show(p,figsize=4)
```

## Exercise 4

*Consider $f(x) = e^{-x^2}\cos 2x$.*

i. *Plot the graph of the function along with Taylors polynomial of degree $1$, $2$, $3$, $6$, $7$, $9$, 1$0$.*

ii. *Use sage interact to create interactive plot to plot Taylors polynomial along with the curve of desired degree at a desired point.*

iii. *Tablulate error term of the approximation of $f$ at $x = 1.2$ when it is appromixated by Taylors polynomial of degree $1$ to $10$ about $x = 1$.*

```python
x, n = var('x n')

f = exp(-x**2)*cos(2*x)
```

```python
# i

a = 1

N = [1, 2, 3, 6, 7, 9, 10]

p=f.plot(-2,3,figsize=4)
for n in N:
    tn(x) = f.taylor(x,a,n) 
    ptn=tn.plot(-2,3,color='red')
    p+=ptn
    
show(p,figsize=8,ymax=1.2,ymin=-1)
```

```python
# ii

@interact
def _(order=(1,(1,10)),x0=(0,(-4,4,0.1))):
    tn = f.taylor(x,x0,order)
    p = plot(f,-4,4, thickness=1)
    dot = point((x0,f(x=x0)),pointsize=20,rgbcolor=(1,0,0))
    pt = plot(tn,x0-1,x0+1, color='green', thickness=1)
    show(dot + p + pt, ymin = -1,ymax=1)
```

```python
# iii

x0 = 1.2
a = 1


errors = [['n', 'error']]

for n in range(1,11):
    tn = f.taylor(x,a,n).substitute(x=x0).n()
    error = f(x0)-tn
    errors.append([n,error])
    
table(rows=errors)
```
