## Exercise 1

*Find the following limits (plot graph as for each function):*

i. $\lim _{x \rightarrow 0}\left(x^{2}-\frac{2^{x}}{1000}\right)$

ii.$\lim _{x \rightarrow 0.5^{-}} \frac{2 x-1}{2 x^{3}-x^{2}}$

iii. $\lim _{x \rightarrow \infty} \frac{x}{x-2^{x}-\sin x }$

```python
# i.

x = var('x')

f = x**2-2**x/1000

p = plot(f(x), (x,-0.1, 0.1))

show(p)
```

```python
show(LatexExpr(r'\lim _{x \rightarrow 0}\left(x^{2}-\frac{2^{x}}{1000}\right) = '), limit(f, x=0))
```

```python
# ii.

x = var('x')

f = x/(2*x**3-x**2)

p = plot(f(x), (x,-0.55, -0.45))

show(p)
```

```python
show(LatexExpr(r'\lim _{x \rightarrow 0.5^{-}} \frac{2 x-1}{2 x^{3}-x^{2}} = '), limit(f, x=-1/2, dir='left'))
```

```python
# iii.

x = var('x')

f = (2*x-1)/(x-2**x-sin(x))

p = plot(f(x), (x,-pi, 4*pi))

show(p)
```

```python
show(LatexExpr(r'\lim _{x \rightarrow \infty} \frac{x}{x-2^{x}-\sin x } = '), limit(f, x='oo'))
```

## Exercise 2

*Find the limit of the following sequences :*

i. $\lim _{n \rightarrow \infty} \frac{\sin n-\cos n}{n}$

ii. $\lim _{n \rightarrow \infty} \frac{n^{3}}{n^{4}+10^{4}}$

iii. $\lim _{n \rightarrow \infty} \left(\frac{1}{n+1}+\frac{1}{n+2}+\cdots+\frac{1}{2 n}\right)$

iv. $\lim _{n \rightarrow \infty} \left(\frac{1}{n^{2}+1}+\frac{1}{n^{2}+2}+\cdots+\frac{1}{n^{2}+n}\right)$

v. $a_{1}=1, \quad a_{n+1}=1+\frac{1}{1+a_{n}}, n>1$

```python
# i.

n = var('n')

u = (sin(n)-cos(n))/n

show(LatexExpr(r'\lim _{n \rightarrow \infty} \frac{\sin n-\cos n}{n} = '), limit(u, n='oo'))
```

```python
# ii.

n = var('n')

u = n**3/(n**4+1000)

show(LatexExpr(r'\lim \frac{n^{3}}{n^{4}+10^{4}} = '), limit(u, n='oo'))
```

```python
# iii.
from sage.calculus.calculus import symbolic_sum

n, k = var('n k')

u = symbolic_sum(k+n, k, 1, n)

show(LatexExpr(r'\lim _{n \rightarrow \infty} \left(\frac{1}{n+1}+\frac{1}{n+2}+\cdots+\frac{1}{2 n}\right) = '), limit(u, n='oo'))
```

```python
# iv.
from sage.calculus.calculus import symbolic_sum

n, k = var('n k')

u = symbolic_sum(1/(n**2+k),  k, 1, n)

show(LatexExpr(r'\lim _{n \rightarrow \infty} \left(\frac{1}{n^2+1}+\frac{1}{n^2+2}+\cdots+\frac{1}{n^2+n}\right) = '), limit(u, n='oo'))
```

```python
# v.

def Y(n, D={}):
    if n in D.keys():
        return D[n]
    else:
        if(n == 1):
            result = 1
        else:
            result = 1+1/(1+Y(n-1))
    D[n] = result
    return result

N = 20
pts = [(n,Y(n)) for n in range(1,N)]

points(pts,figsize=4).show()

sol = solve(x==1+1/(1+x),x)
```
