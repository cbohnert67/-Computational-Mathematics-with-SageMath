## Exercise 1

*Plot the graph of figure eight given by $x = \sin t$ and $y = \sin 2t$ for $0 \leq t \leq 2Ï€$.*

```python
t = var('t')

eqns = [sin(t), sin(2*t)]

p = parametric_plot(eqns, (t, 0, 2*pi), color='red', thickness=3)
txt = text('Figure of eight',(-0.7, 1.2), fontsize=14)
show(p+txt)
```

## Exercise 2

*Suppose $\pi(n)$ is the number of primes less than or equal to the natural number $n$. Plot graphs of $\pi(n)$, $\frac{\pi(n)}{n}$ and  $\frac{\pi(n)}{n / \log (n)}$. What do you abserve?*

```python
n = var('n')

f1(n) = prime_pi(n)
f2(n) = prime_pi(n)/n
f3(n) = prime_pi(n)/(n/log(n))

p1 = plot(f1(n), (n, 2, 1000), color='red', thickness=2)
txt1 = text(r'$\pi(n)$', (200,150), fontsize=14)
p2 = plot(f2(n), (n, 2, 1000), ymin=0, color='red', thickness=2)
txt2 = text(r'$\frac{\pi(n)}{n}$', (200,0.5), fontsize=14)
p3 = plot(f1(n), (n, 2, 1000), color='red', thickness=2)
txt3 = text(r'$\frac{\pi(n)}{n / \log (n)}$', (200,150), fontsize=14)

L = [p1+txt1,p2+txt2,p3+txt3]
G=graphics_array(L,1)
G.show(figsize=(10,4))
```

## Exercise 3

*Plot the unit circle and inside this plot regular n-gons for $n = 3, 4,5,6$. Think of plotting for any $n$.*

```python
n = 5

G = Graphics()
c = circle((0,0),1, color='red', thickness=1)
ngon_points = [[cos(2*pi*k/n), sin(2*pi*k/n)] for k in range(n)]
ngon =  polygon2d(ngon_points,  fill=False, thickness=1)
G += c
G += ngon
G.axes(show=False)
G.show()
```

## Exercise 4

*Plot the graph of the function $f(x,y)=10xy e^{-(x^2+y^2)}$ for $-1 \leq x, y \leq 2$. Also plot the contours in the same domain.*

```python
x, y = var('x y')

f(x,y) = 10*x*y*exp(-x**2-y**2)

p = plot3d(f(x,y), (x, -2,2), (y, -2,2))
cp = contour_plot(f(x,y),(x,-2,2),(y,-2,2), cmap='Blues')
show(p)
show(cp)
```

## Exercise 5

*Plot the surface of revolution about z-axis of the area between the curves $y = x$ and $y = x^2$ between $x = 0$ to $x = 1$.*

```python
x = var('x')

curve1 = x
curve2 = x**2


p1 = revolution_plot3d(curve1, (x,0,1),(0, 2*pi),show_curve=True, parallel_axis='z', color='blue',opacity=0.7)
p2 = revolution_plot3d(curve2, (x,0,1),(0, 2*pi),show_curve=True, parallel_axis='z', color='blue',opacity=0.5)

show(p1+p2)
```
